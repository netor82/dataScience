### Probabilidad conjunta
probabil√≠stico frecuentista. $p\left(Y=y_{j},X=x_{i}\right)=p\left(X=x_{i},Y=y_{j}\right)=\frac{n_{i,j}}{N} (1)$

### Regla de la suma

La probabilidad de que $X=x_{i}$ sin importar el valor de $Y$ viene dado por: $p\left(X=x_{i}\right)=p\left(X=x_{i},Y=y_{1},\ldots,y_{L}\right)=\sum_{j=1}^{L}p\left(X=x_{i},Y=y_{j}\right)=\frac{c_{i}}{N}(2)$

A $p\left(X=x_{i}\right)$ tambi√©n se le conoce como probabilidad marginal de $X$

### Probabilidad condicional
$p\left(Y=y_{j}|X=x_{i}\right)=\frac{n_{i,j}}{c_{i}}(3)$.

### Regla del producto de probabilidad
* Regla de la suma: $p\left(X\right)=\sum_\limits{Y}p\left(X,Y\right)$.
* Regla del producto: $p\left(X,Y\right)=p\left(Y|X\right)\cdot p\left(X\right)$.

## Teorema de Bayes
$$
\Rightarrow p\left(Y|X\right)=\frac{p\left(X|Y\right)\cdot p\left(Y\right)}{p\left(X\right)}
$$

$
p\left(H|D\right)=\frac{p\left(D|H\right)\cdot p(H)}{p(D)}
$

  * **Verosimilitud (likelihood)**: Es la probabilidad $p(D|H)$ de observar la evidencia $D=d$ dado que la hip√≥tesis $H=h$ es verdadera.

  * **Probabilidad a priori o marginal**: Se refiere a la probabilidad $p\left(H=h\right)$ de que una hip√≥tesis $H=h$ se de antes de observar uno o m√°s datos $D=d$ al que puede estar condicionada la hip√≥tesis.

  * **Evidencia**: Es la probabilidad $p(D=d)$ de obtener la evidencia $D=d$

  * **Probabilidad a posteriori:**
  Corresponde a la probabilidad condicional 
  $p\left(H=h|D=d\right)$
  la cual describe la probabilidad de que la hip√≥tesis $H=h$ se de, dados los datos $D=d$.

## Funci√≥n de distribuci√≥n de probabilidad acumulativa (CDF) - Clase 1, grabaci√≥n Lecc_2

Indica la probabilidad de que una variable aleatoria $X$ tome un valor menor o igual a un n√∫mero real $x$. En otras palabras $F_X(x) = p(X \leq x)$.

Para simplificar la notaci√≥n $F_X(x)$ como $F(x)$.

Propiedades:
1. $F(-\infty)=0$

2. $F(\infty)=1$

3. $0 \leq F(x) \leq 1$

4. $ F(x_1) \leq F(x_2)$ si $x_1 \leq x_2$

5. $p(x_1 \leq X \leq x_2) = F(x_2) - F(x_1)$


## Funci√≥n de densidad de probabilidad (PDF)
$$f(x)= \frac{dF(x)}{dx}$$

Para variable aleatoria continua: 

1. $f\left(x\right)\geq 0$

2. $\int_\limits{-\infty}^{\infty}f\left(x\right)\textrm{d}x=1$

3. $F(x) = \int_\limits{-\infty}^{x}f\left(u\right)\textrm{d}u$

4. $p(x_1 \leq X \leq x_2) = \int_\limits{x_1}^{x_2}f\left(x\right)\textrm{d}x.$

### funci√≥n de masa de probabilidad
Para discretas:
1. $f\left[x\right]\geq 0$
2. $\sum \limits_x f\left[x\right]=1$
3. $F[x] = \sum_\limits{x_i \leq x}^{}f\left[x_i\right]$
4. $p[x_1 \leq X \leq x_2] = \sum_\limits{x_1 \leq x_i \leq x_2}^{}f\left[x_i\right].$

## Esperanza, valor esperado o media
$
\mu_{X}=\mathbb{E}\left(X\right)=\int \limits_{-\infty}^{\infty} x\,f\left(x\right)\textrm{d}x
$

$\mu_{X}=\mathbb{E}\left[X\right]=\sum \limits_x x\,f\left[x\right]$

Si n√∫mero finito de $N$ observaciones **aproximar** la esperanza como
$
\mathbb{E}\left[X\right]\cong\frac{1}{N}\sum \limits _{i=1}^{N}x_i

$\mathbb{E}\left[a\right]=a$
* Homogeneidad: 
$\mathbb{E}\left[a\,X\right]=a\,\mathbb{E}\left[X\right].$
*Superposici√≥n: 
$\mathbb{E}\left[g\left(X\right)+h\left(X\right)\right]=\mathbb{E}\left[g\left(X\right)\right]+\mathbb{E}\left[h\left(X\right)\right].$

## Varianza - Clase 2, grabaci√≥n Lecc_2
Concentraci√≥n o dispersi√≥n alrededor de la media

$$
\sigma_{X}^{2}\textrm{=var}\left[X\right]=\mathbb{E}\left[\left(X-\mathbb{E}\left[X\right]\right)^{2}\right]
$$

Propiedades : 

* Si $a$ es un escalar $a\in\mathbb{R}$, se tiene que: $\textrm{Var}\left[a\right]=0$.
*Multiplicaci√≥n por escalar de la entrada: 
$\textrm{Var}\left[a\,X\right]=a^{2}\textrm{Var}\left[X\right]$, $\forall a\in\mathbb{R}$.

Si √∫nicamente se tiene un estimado de la media $\hat \mu$, y se toma la suposici√≥n de que los datos provienen de una distribuci√≥n normal, la varianza **se estima**:

\begin{equation}
\sigma_{X}^{2}\cong\frac{1}{N-1}\sum_{i=1}^{N}\left(x_i-\hat \mu_{X}\right)^{2}. (1)
\end{equation}

El estimador anterior se dice que no es "sesgado" si la media real no es conocida. Si la media real de la poblaci√≥n es conocida, o bien se tiene un N muy grande, se normaliza por $N$



### Covarianza
Mide la variaci√≥n conjunta de tales variables aleatorias


$$
\Sigma_{X,Y}=\textrm{cov}\left(X,Y\right)=\mathbb{E}\left[\left(X-\mathbb{E}\left[X\right]\right)\left(Y-\mathbb{E}\left[Y\right]\right)\right]
$$

Con $N$ observaciones de ambas variables, se **estimada** por: 
$$
\Sigma_{X,Y}\cong\frac{1}{N-1}\sum_{i=1}^{N}\left(x_i-\overline{\mu}_{X}\right)\left(y_i-\overline{\mu}_{Y}\right)
$$

La matriz de covarianza se puede reescribir como: 

\begin{equation}
\Sigma=\begin{bmatrix}\sigma_{X_{1}}^{2} & \ldots & \mathbb{E}\left[\left(X_{1}-\mathbb{E}\left[X_{1}\right]\right)\left(X_{n}-\mathbb{E}\left[X_{n}\right]\right)\right]\\
\vdots & \ddots & \vdots\\
\mathbb{E}\left[\left(X_{n}-\mathbb{E}\left[X_{n}\right]\right)\left(X_{1}-\mathbb{E}\left[X_{1}\right]\right)\right] & \ldots & \sigma_{X_{n}}^{2}
\end{bmatrix}
\end{equation}

## Binomial
√©xito o fracaso - prueba de Bernoulli o ensayo de Bernoulli

- Consiste en un n√∫mero fijo, $n$, de pruebas id√©nticas.
- Cada prueba resulta en uno de dos resultados: √©xito o fracaso.
- Las pruebas son independientes.
- La probabilidad de √©xito en una sola prueba es igual a  $p$ y es el mismo de una prueba a otra. La probabilidad de fracaso es igual a $q=(1-p)$. Se puede describir ambas probabilidades con un √∫nico par√°metro.
- La variable aleatoria de inter√©s $X$, es discreta y describe el n√∫mero de pruebas exitosas observado durante las $n$ pruebas.



N√≥tese que los eventos que se pueden modelar no requieren tener s√≥lo dos valores expl√≠citamente, sino que se pueden agrupar en 2 categor√≠as. Por ejemplo, en un examen de selecci√≥n m√∫ltiple con $n$ opciones, lo importante es modelar si la respuesta es correcta o no (no la cantidad de opciones).

## Geom√©trica
N√∫mero de pruebas para que ocurra el primer √©xito. $x$ pruebas para encontrar la primera ocurrencia del evento.

Duraci√≥n de tiempos de espera

- pruebas id√©nticas e independientes
- con dos posibles resultados: √©xito o fracaso
- probabilidad de √©xito $p$ constante de una prueba a otra.

\begin{equation}
p(x)=(1-p)^{x-1}p,\quad x=1,2,\ldots, \quad 0\leq p \leq 1
\end{equation}

La media y varianza est√°n dadas por:

\begin{equation}
\mu=\mathrm{E}(X)=\dfrac{1}{p}
\end{equation}

\begin{equation}
\sigma^2=\mathrm{Var}(X)=\dfrac{1-p}{p^2}
\end{equation}

## Poisson
 sirve para describir la probabilidad de llegada de nuevos clientes a una cola en un intervalo de tiempo o la probabilidad de un accidente automovil√≠stico en un tramo de carretera durante una semana. Clientes por unidad de tiempo.

- El total de eventos en cada segmento es independiente de otros segmentos.

\begin{equation}
p(x)=\dfrac{\lambda^x}{x!}e^{-\lambda} \space \space x=0, 1, 2,\ldots \space \lambda>0
\end{equation}

## Exponencial

f(x)=\left\lbrace
\begin{array}{ll}
\\ \lambda\space e^{-\lambda x}   &x \geq 0, \space \lambda \geq 0\\
\\
 0& \text{cualquier otro punto}

\right.
\end{equation}

La esperanza y varianza est√°n dadas por:

\begin{equation}

\end{equation}
