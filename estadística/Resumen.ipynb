Dos ramas o enfoques, la teoría de probabilidades **frecuentista**, basada en la frecuencia de ocurrencias de eventos, y la
**bayesiana**, que define las probabilidades de acuerdo a suposiciones previas sobre el fenómeno.

- espacio muestral $\Omega$
- experimiento $\omega \in \Omega$
- espacio de eventos $\mathcal{F}$. Conjunto de eventos $A$ ⊆ Ω. Incluye vacío $\mathcal{F}:\emptyset \in \mathcal{F}$

Ejemplo de dados:

$\Omega=\left\{ 1,2,3,4,5,6\right\} $ con $\omega_{1}=1,\omega_{2}=2,\ldots$ etc. El espacio de eventos más simple es $\mathcal{F}=\left\{ \emptyset,\Omega\right\} $

* $p(\emptyset) = 0$
* $p(\Omega) =1$
* $0 \leq p(A) \leq 1, \forall A \in \mathcal{F}$
* Si $ A_1, A_2, ..., A_n $ son eventos disjuntos tal que $A_i \cap A_j = \emptyset$, si $ i \neq j $ entonces se tiene que: $p(\bigcup\limits_{i=0}^{n} A_i) = \sum\limits_{i=0}^{n} p(A_i) $

Ejemplo de dados:

Otro espacio de eventos posible  $\mathcal{F}$ es el conjunto de todos los subconjuntos de $\Omega$. Para este último espacio de eventos, se puede asignar la probabilidad de cada conjunto $A_{i}$ en tal espacio de eventos $\mathcal{F}$ como $\frac{k}{n}$donde $k$ es la cantidad de elementos $\left|A_{i}\right|$ o cardinalidad y $n=\left|\Omega\right|=6$.  Por ejemplo $p\left(A_{1}=\left\{ 1,2,3,4\right\} \right)=\frac{4}{6}$


### Propiedades de los eventos
* Si $A\subseteq B\Rightarrow p\left(A\right)\leq p\left(B\right)$
* Cota de la intersección $p\left(A\cap B\right)\leq\min\left(p\left(A\right),p\left(B\right)\right)$
*Cota de la unión $p\left(A\cup B\right)\leq p(A)+p(B)$
*Complemento $p\left(\Omega-A\right)=1-p(A)$

## Variables aleatorias
variable aleatoria $X$ o $X\left(\omega\right)$ es una función que asigna un número real a cada posible resultado $w$ de un experimento aleatorio, $X:\Omega\rightarrow\mathbb{R}$.

- variable **discreta** tome un valor $x_i$ viene dado por $ p\left(X=x_i\right)$.  Lanzar un dado, número de personas, edad
- **continua** la probabilidad de que la variable aleatoria tome un valor en el intervalo $[a,b]$ $/$ $a\in\mathbb{R} \land b\in\mathbb{R}$ como $p\left(a\leq X\leq b\right)$.  Tiempo, corriente eléctrica, tempreatura.

### Probabilidad conjunta
probabilístico frecuentista. $p\left(Y=y_{j},X=x_{i}\right)=p\left(X=x_{i},Y=y_{j}\right)=\frac{n_{i,j}}{N} (1)$

### Regla de la suma

La probabilidad de que $X=x_{i}$ sin importar el valor de $Y$ viene dado por: $p\left(X=x_{i}\right)=p\left(X=x_{i},Y=y_{1},\ldots,y_{L}\right)=\sum_{j=1}^{L}p\left(X=x_{i},Y=y_{j}\right)=\frac{c_{i}}{N}(2)$

A $p\left(X=x_{i}\right)$ también se le conoce como probabilidad marginal de $X$

### Probabilidad condicional
$p\left(Y=y_{j}|X=x_{i}\right)=\frac{n_{i,j}}{c_{i}}(3)$.

### Regla del producto de probabilidad
* Regla de la suma: $p\left(X\right)=\sum_\limits{Y}p\left(X,Y\right)$.
* Regla del producto: $p\left(X,Y\right)=p\left(Y|X\right)\cdot p\left(X\right)$.

## Teorema de Bayes
$$
\Rightarrow p\left(Y|X\right)=\frac{p\left(X|Y\right)\cdot p\left(Y\right)}{p\left(X\right)}
$$

$
p\left(H|D\right)=\frac{p\left(D|H\right)\cdot p(H)}{p(D)}
$

  * **Verosimilitud (likelihood)**: Es la probabilidad $p(D|H)$ de observar la evidencia $D=d$ dado que la hipótesis $H=h$ es verdadera.

  * **Probabilidad a priori o marginal**: Se refiere a la probabilidad $p\left(H=h\right)$ de que una hipótesis $H=h$ se de antes de observar uno o más datos $D=d$ al que puede estar condicionada la hipótesis.

  * **Evidencia**: Es la probabilidad $p(D=d)$ de obtener la evidencia $D=d$

  * **Probabilidad a posteriori:**
  Corresponde a la probabilidad condicional 
  $p\left(H=h|D=d\right)$
  la cual describe la probabilidad de que la hipótesis $H=h$ se de, dados los datos $D=d$.

## Función de distribución de probabilidad acumulativa (CDF) - Clase 1, grabación Lecc_2

Indica la probabilidad de que una variable aleatoria $X$ tome un valor menor o igual a un número real $x$. En otras palabras $F_X(x) = p(X \leq x)$.

Para simplificar la notación $F_X(x)$ como $F(x)$.

Propiedades:
1. $F(-\infty)=0$

2. $F(\infty)=1$

3. $0 \leq F(x) \leq 1$

4. $ F(x_1) \leq F(x_2)$ si $x_1 \leq x_2$

5. $p(x_1 \leq X \leq x_2) = F(x_2) - F(x_1)$


## Función de densidad de probabilidad (PDF)
$$f(x)= \frac{dF(x)}{dx}$$

![](https://www.vosesoftware.com/riskwiki/images/image15_400.gif)

Para variable aleatoria continua: 

1. $f\left(x\right)\geq 0$

2. $\int_\limits{-\infty}^{\infty}f\left(x\right)\textrm{d}x=1$

3. $F(x) = \int_\limits{-\infty}^{x}f\left(u\right)\textrm{d}u$

4. $p(x_1 \leq X \leq x_2) = \int_\limits{x_1}^{x_2}f\left(x\right)\textrm{d}x.$

### función de masa de probabilidad
Para discretas:
1. $f\left[x\right]\geq 0$
2. $\sum \limits_x f\left[x\right]=1$
3. $F[x] = \sum_\limits{x_i \leq x}^{}f\left[x_i\right]$
4. $p[x_1 \leq X \leq x_2] = \sum_\limits{x_1 \leq x_i \leq x_2}^{}f\left[x_i\right].$

## Esperanza, valor esperado o media
$
\mu_{X}=\mathbb{E}\left(X\right)=\int \limits_{-\infty}^{\infty} x\,f\left(x\right)\textrm{d}x
$

$\mu_{X}=\mathbb{E}\left[X\right]=\sum \limits_x x\,f\left[x\right]$

Si número finito de $N$ observaciones **aproximar** la esperanza como
$
\mathbb{E}\left[X\right]\cong\frac{1}{N}\sum \limits _{i=1}^{N}x_i
$

Propiedades de la esperanza: 

* Si $a$ es un escalar, tal que $a\in\mathbb{R}$, se tiene que: 
$\mathbb{E}\left[a\right]=a$
* Homogeneidad: 
$\mathbb{E}\left[a\,X\right]=a\,\mathbb{E}\left[X\right].$
*Superposición: 
$\mathbb{E}\left[g\left(X\right)+h\left(X\right)\right]=\mathbb{E}\left[g\left(X\right)\right]+\mathbb{E}\left[h\left(X\right)\right].$

## Varianza - Clase 2, grabación Lecc_2
Concentración o dispersión alrededor de la media

$$
\sigma_{X}^{2}\textrm{=var}\left[X\right]=\mathbb{E}\left[\left(X-\mathbb{E}\left[X\right]\right)^{2}\right]
$$

Propiedades : 

* Si $a$ es un escalar $a\in\mathbb{R}$, se tiene que: $\textrm{Var}\left[a\right]=0$.
*Multiplicación por escalar de la entrada: 
$\textrm{Var}\left[a\,X\right]=a^{2}\textrm{Var}\left[X\right]$, $\forall a\in\mathbb{R}$.

Si únicamente se tiene un estimado de la media $\hat \mu$, y se toma la suposición de que los datos provienen de una distribución normal, la varianza **se estima**:

\begin{equation}
\sigma_{X}^{2}\cong\frac{1}{N-1}\sum_{i=1}^{N}\left(x_i-\hat \mu_{X}\right)^{2}. (1)
\end{equation}

El estimador anterior se dice que no es "sesgado" si la media real no es conocida. Si la media real de la población es conocida, o bien se tiene un N muy grande, se normaliza por $N$



### Covarianza
Mide la variación conjunta de tales variables aleatorias


$$
\Sigma_{X,Y}=\textrm{cov}\left(X,Y\right)=\mathbb{E}\left[\left(X-\mathbb{E}\left[X\right]\right)\left(Y-\mathbb{E}\left[Y\right]\right)\right]
$$

Con $N$ observaciones de ambas variables, se **estimada** por: 
$$
\Sigma_{X,Y}\cong\frac{1}{N-1}\sum_{i=1}^{N}\left(x_i-\overline{\mu}_{X}\right)\left(y_i-\overline{\mu}_{Y}\right)
$$

La matriz de covarianza se puede reescribir como: 

\begin{equation}
\Sigma=\begin{bmatrix}\sigma_{X_{1}}^{2} & \ldots & \mathbb{E}\left[\left(X_{1}-\mathbb{E}\left[X_{1}\right]\right)\left(X_{n}-\mathbb{E}\left[X_{n}\right]\right)\right]\\
\vdots & \ddots & \vdots\\
\mathbb{E}\left[\left(X_{n}-\mathbb{E}\left[X_{n}\right]\right)\left(X_{1}-\mathbb{E}\left[X_{1}\right]\right)\right] & \ldots & \sigma_{X_{n}}^{2}
\end{bmatrix}
\end{equation}

## Binomial
éxito o fracaso - prueba de Bernoulli o ensayo de Bernoulli

- Consiste en un número fijo, $n$, de pruebas idénticas.
- Cada prueba resulta en uno de dos resultados: éxito o fracaso.
- Las pruebas son independientes.
- La probabilidad de éxito en una sola prueba es igual a  $p$ y es el mismo de una prueba a otra. La probabilidad de fracaso es igual a $q=(1-p)$. Se puede describir ambas probabilidades con un único parámetro.
- La variable aleatoria de interés $X$, es discreta y describe el número de pruebas exitosas observado durante las $n$ pruebas.



Nótese que los eventos que se pueden modelar no requieren tener sólo dos valores explícitamente, sino que se pueden agrupar en 2 categorías. Por ejemplo, en un examen de selección múltiple con $n$ opciones, lo importante es modelar si la respuesta es correcta o no (no la cantidad de opciones).

## Geométrica
Número de pruebas para que ocurra el primer éxito. $x$ pruebas para encontrar la primera ocurrencia del evento.

Duración de tiempos de espera

- pruebas idénticas e independientes
- con dos posibles resultados: éxito o fracaso
- probabilidad de éxito $p$ constante de una prueba a otra.

\begin{equation}
p(x)=(1-p)^{x-1}p,\quad x=1,2,\ldots, \quad 0\leq p \leq 1
\end{equation}

La media y varianza están dadas por:

\begin{equation}
\mu=\mathrm{E}(X)=\dfrac{1}{p}
\end{equation}

\begin{equation}
\sigma^2=\mathrm{Var}(X)=\dfrac{1-p}{p^2}
\end{equation}


## Combinatoria

## Función de probabilidad
cuántas diferentes maneras se podrían dar las pruebas de forma que tengamos $x$ éxitos, de un total de $n$.

\begin{equation}
p(x) = {n \choose x} p^x(1-p)^{n-x},\quad x=0,1,\ldots,n,\quad 0\leq p \leq 1
\end{equation}

La esperanza y varianza están dadas únicamente por los dos parámetros $n$ y $p$

\begin{equation}
\mu=\mathrm{E}(X)=np
\end{equation}

\begin{equation}
\sigma^2=\mathrm{Var}(X)=np(1-p)
\end{equation}

- Eficacia de un tratamiento en la cura de una enfermedad

## Poisson
 sirve para describir la probabilidad de llegada de nuevos clientes a una cola en un intervalo de tiempo o la probabilidad de un accidente automovilístico en un tramo de carretera durante una semana. Clientes por unidad de tiempo.
 
$X$ se interpreta como el número de eventos en el intervalo

**cuantas ocurrencias de un evento se dan por segmento** / conteo con unidades

- $p$ es la misma en en cualquiera de los intervalos y es proporcional al largo del segmento.
- El total de eventos en cada segmento es independiente de otros segmentos.

\begin{equation}
p(x)=\dfrac{\lambda^x}{x!}e^{-\lambda} \space \space x=0, 1, 2,\ldots \space \lambda>0
\end{equation}

y su esperanza y varianza son: 


\begin{equation}
\mu=\mathrm{E}(X)= \lambda
\end{equation}

\begin{equation}
\sigma^2=\mathrm{Var}(X)=\lambda
\end{equation}

## Exponencial
En cierto tiempo suceda un determinado evento. Distancia (tiempo) entre eventos de un proceso Poisson (proceso en el cual los eventos **ocurren de forma continua e independiente y a una tasa promedio constante**)

\begin{equation}
f(x)=\left\lbrace
\begin{array}{ll}
\\ \lambda\space e^{-\lambda x}   &x \geq 0, \space \lambda \geq 0\\
\\
 0& \text{cualquier otro punto}
\end{array}
\right.
\end{equation}

La esperanza y varianza están dadas por:

\begin{equation}
\mu=\mathrm{E}(X)= \dfrac{1}{\lambda}
\end{equation}

\begin{equation}
\sigma^2=\mathrm{Var}(X)=\dfrac{1}{\lambda^2}
\end{equation}
